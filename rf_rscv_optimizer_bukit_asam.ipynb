#### Splitting dataset

```python
X = df.iloc[:,2:5]
Y = df.iloc[:,5:6]

print('Data X/Independen\n',X)
print('Data Y/Dependen\n',Y)
```

#### Normalisasi data

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0,1))

scaled_X = min_max_scaler.fit_transform(X)
scaled_Y = min_max_scaler.fit_transform(Y)

print('Data X setelah dinormalisasi\n',scaled_X)
print('Data Y setelah dinormalisasi\n',scaled_Y)
```

```python
scaled_X = np.array(scaled_X)
scaled_Y = np.array(scaled_Y)
```

```python
scaled_X
```

```python
scaled_Y
```

#### Train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(scaled_X,scaled_Y,test_size=0.30,random_state=40)
```

#### Random Search Cross Validation (CV) Optimizer

```python
from sklearn.model_selection import RandomizedSearchCV

# Menghitung jumlah pohon dalam RF
n_estimators = [int(x) for x in np.linspace(start=200, stop=2000,num=10)]

# Jumlah fitur yang dipertimbangkan di setiap split
max_features = ['auto','sqrt']

# Jumlah level pada pohon
max_depth = [int(x) for x in np.linspace(10,110,num=110)]
max_depth.append(None)

# Jumlah minimal sample yang dibutuhkan untuk split node
min_samples_split = [2,5,10]

# Jumlah minimal sample yang dibutuhkan untuk setiap node leaf (daun)
min_samples_leaf = [1,2,4]

# Metode pemilihan sample untuk setiap pohon training
bootstrap = [True, False]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

#### Melatih model Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(random_state = 40)

rf_random = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, n_iter=100, cv=3, verbose=2, random_state=40, n_jobs=-1)

# Fitting model
rf_random.fit(X_train, Y_train)
```

```python
rf_random.best_params_
```

```python
rf.fit(X_train,Y_train)
```

```python
rf_prediction = rf.predict(X_test)

rf_prediction
```

#### Evaluasi model

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

mae = mean_absolute_error(Y_test,rf_prediction)
mse = mean_squared_error(Y_test,rf_prediction)

print('Nilai MAE RF/RSCV : ',mae)
print('Nilai MSE RF/RSCV : ',mse)
```

```python
from math import sqrt

rmse = sqrt(mse)
print('Nilai RMSE RF/RSCV : ',rmse)
```

```python
from sklearn.metrics import r2_score
print('Hasil R2 Score RF/RSCV : ',r2_score(Y_test,rf_prediction)*100)
```

#### Tes prediksi RF/RSCV (2)

```python
rf_prediction_dua = rf.predict(scaled_X)

rf_prediction_dua
```

```python
rf_prediction_dua = rf_prediction_dua.reshape(1,-1)

rf_prediction_dua
```

#### Denormalisasi nilai prediksi RF/RSCV

```python
inv_rf_prediction_dua = min_max_scaler.inverse_transform(rf_prediction_dua)

inv_rf_prediction_dua = inv_rf_prediction_dua.reshape(-1)

inv_rf_prediction_dua
```

```python
df_rf_prediction_dua = pd.DataFrame({'Prediction RF/RSCV':inv_rf_prediction_dua})
```

```python
df_rf_prediction_dua
```

#### Join tabel ke dataset utama

```python
df = df.merge(df_rf_prediction_dua,how='left',left_index=True,right_index=True)

df
```
